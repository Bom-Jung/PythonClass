```python
class Counter:
  number = 0
  def __init__(self, number=0):
    if number >= 100 or number <= -1:
      number = 0
    else:
      self.number += number

  def reset(self):
    self.number = 0

  def inc(self):
    if self.number+ 1 >=100:
      self.number = 0
    else:
      self.number +=1

  def dec(self):
    if self.number -1 <= -1:
      self.number = 0
    else:
      self.number += -1

  def count_print(self):
    print("number의 값은 %d 입니다" %(self.number))

c1=Counter()
c2=Counter()
c1.inc()
c2.inc()
c1.inc()
c1.count_print()
c2.count_print()
```

```python
class BankAccount:
  name = ''
  account_num = ''
  balance = 0
  def __init__ (self,name,account_num,balance=0):
    self.name = name
    self.account_num = account_num
    self.balance = balance

  def info_print(self): 
    print("계좌주인은 %s 계좌번호는 %s 잔액은 %s 입니다." %(self.name,self.account_num,self.balance))

  def get_name(self):
    return self.name

  def get_account_num(self):
    return self.account_num

  def get_balance(self):
    return self.balance

  def deposit(self,money):
    self.balance += money

  def withdraw(self,money):
    if self.balance < money :
      print("잔액부족, 계좌 잔액 : %s , 인출 요구 금액 %d" % (self.balance,money))
    else:
      self.balance -= money

my_account= BankAccount('홍길동','1234-0001')
my_account.info_print()
my_account.deposit(2000)
my_account.info_print()
my_account.withdraw(500)
my_account.info_print()
my_account.withdraw(5000)
```
